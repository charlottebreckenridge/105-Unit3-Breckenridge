To save time attempting to cull various columns at once, the phrase "iloc" is very useful. It goes before our brackets and lets pandas accept two numbers separated by a comma. 

The first number distinguishes the row and the second distinguishes the column.

It appears as so:

df.iloc[2,3]

or if you wish to use every  variable in column three use a semicolon -

df.iloc[:,3]


```python
df.iloc[:,3]
```

Pandas also allows you to filter the data by using logical conditions based on true/false.

This might look like so -

```python
df[df["State"] == "FL"]
```

As shown above, when frames are combined the result is all data contained in rows where the state equals FL. 

To use this FL data frame later on, we can create an "RI_subset" which would appear as follows. 

RI_subset = df[df["State"] == "FL"].copy()

Be careful not to forget the ".copy()" phrase on the end to avoid only an exact copy of the original data frame. 

```python
RI_subset = df[df["State"] == "FL"].copy()
```

To export subsets as new and analyzable csv's, add this phrase ".to_csv()" to the end of your new RI subset as follows - 

RI_subset.to_csv("RI_subset.csv", index=False)

To eliminate all the rows from the original panda csv file, including "index=false" to the statement will let it know not to index those numbers. 

```python
RI_subset.to_csv("RI_subset.csv", index=False)
```

Now we should have a new csv file titled "RI_subset.csv" exactly in our folder, containing data for every item with the state marked FL.
